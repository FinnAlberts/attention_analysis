### 10. Overview

```python
import inspect, sys

exclude = ['BaseTimeSeriesDataset', 'MultivarPeriodicDataset', 'PeriodicSyntheticDataset', 'SimplePeriodicDataset']

module_dict = inspect.getmembers(sys.modules[SD.__name__])

step_signals = [obj for name, obj in module_dict 
                if inspect.isclass(obj) and 'Step' in name]

step_signals[0], step_signals[2] = step_signals[2], step_signals[0]
step_signals[1], step_signals[2] = step_signals[2], step_signals[1]

sine_signals = [obj for name, obj in module_dict 
                if inspect.isclass(obj) and 'Sine' in name]

sine_signals[0], sine_signals[2] = sine_signals[2], sine_signals[0]
sine_signals[1], sine_signals[2] = sine_signals[2], sine_signals[1]

saw_signals = [obj for name, obj in module_dict 
                if inspect.isclass(obj) and 'Saw' in name]

saw_signals[0], saw_signals[2] = saw_signals[2], saw_signals[0]
saw_signals[1], saw_signals[2] = saw_signals[2], saw_signals[1]

fig, axes = plt.subplots(3, 3, figsize=(15, 10), sharex=True)

args = [('', r'$\alpha=0.1$', ''), ('decay', r'$\alpha_1=0.1$', r'$\alpha_2=0.001$'), ('multi', r'$\alpha_1=0.1$', r'$\alpha_2=0.05$')]
COLOR = 'black'

for ax, signal, (prefix, arg1, arg2) in zip(axes[0], step_signals, args):
    data = signal(len_timeseries=len_timeseries, seq_len=seq_len, shift=shift, mode="overlapping")
    ax.plot(data.X, data.fX, COLOR)
    ax.set_title(f"{prefix} step, {arg1}, {arg2}")
    ax.set_xticks([])
    ax.set_yticks([])

for ax, signal, (prefix, arg1, arg2) in zip(axes[1], sine_signals, args):
    data = signal(len_timeseries=len_timeseries, seq_len=seq_len, shift=shift, mode="overlapping")
    ax.plot(data.X, data.fX, COLOR)
    ax.set_title(f"{prefix} sine, {arg1}, {arg2}")
    ax.set_xticks([])
    ax.set_yticks([])
    
for ax, signal, (prefix, arg1, arg2) in zip(axes[2], saw_signals, args):
    data = signal(len_timeseries=len_timeseries, seq_len=seq_len, shift=shift, mode="overlapping")
    ax.plot(data.X, data.fX, COLOR)
    ax.set_title(f"{prefix} sawtooth, $p=5$, {arg1}, {arg2}")
    ax.set_xticks([])
    ax.set_yticks([])
    
plt.tight_layout()

fig.savefig(f"../img/all_synthetic_signals.pdf", dpi=1200)
```

### new approach

```python
from src.Datasets.TrafficDataset import TrafficDataset, TrafficMeta

data = TrafficDataset.load()

# shape: (t, time_series)
data.dates.shape, data.ids.shape, data.values.shape
```

```python
splits = {
    # https://arxiv.org/pdf/1704.04110.pdf
    'deepar': '2008-06-14 23',
    # Figure B.5 http://proceedings.mlr.press/v97/wang19k/wang19k-supp.pdf
    'deepfactors': '2008-01-13 23',
    # the last 7 days according to our holidays assumption
    'last': '2009-03-24 00'
}

fig, axes = plt.subplots(len(splits), 1, figsize=(5*len(splits), 5), sharex=True)

for ax, (name, split) in zip(axes, splits.items()):
    
    train, test = data.split_by_date(split)
    print(f"{name}: \n\ttrain shape: {train.values.shape}, test shape: {test.values.shape}")
    ax.plot(range(train.values.shape[0]), train.values[:, 0], label='train')
    ax.plot(range(train.values.shape[0], train.values.shape[0] + test.values.shape[0]), test.values[:, 0], label='test')
    ax.legend()
    

plt.tight_layout()
```

```python
traffic_lane_idx = 20

one_day_range = (data.dates > '2008-06-14 23') & (data.dates <= '2008-06-15 23')
covars = list(map(lambda x: x.split()[1], data.dates[one_day_range]))

plt.figure(figsize=(20, 5))

plt.plot(covars, data.values[one_day_range, traffic_lane_idx])
```

```python
from src.sampler import TimeseriesSampler

lookback = 7
input_size = TrafficMeta.horizon * lookback

train, test = data.split_by_date(splits['deepar'])

sampler = TimeseriesSampler(timeseries=train.values, 
                            insample_size=input_size, 
                            outsample_size=TrafficMeta.horizon,
                            window_sampling_limit=10*TrafficMeta.horizon,
                            batch_size=32)
```

```python
insample, insample_mask, outsample, outsample_mask = next(iter(sampler))
    
print(insample.shape, insample_mask.shape, outsample.shape, outsample_mask.shape)

for i in range(1):
    plt.plot(range(insample.shape[1]), insample[i, :])
    plt.plot(range(insample.shape[1], insample.shape[1] + outsample.shape[1]), outsample[i, :])
```
